Hence, It more clearly shows that **Cosmetics** Product Category has got the arrival of more Loyal Customers Compared to others and the **Home_decor** has got the least

**FEATURE ENGINEERING**

Let's Check for the null Values in the data first!!

```
df.isnull().sum()
```

Filling NaN values.

```
 #For Stall_no, As no of stall is a categorical feature we will use mode to fill the NaN value.
 df['Stall_no'].mode()   #it gives value 20.0
 df['Stall_no'].fillna(20.0,inplace=True)
```

```
#also For Discount_avail, we will use mode value to fill the NaN value.
df['Discount_avail'].mode()  #it gives value 0
df['Discount_avail'].fillna(0,inplace=True)
```

```
#For charges_1 median is used to fill NaN values.
df['charges_1'].median()  #median=252.0
df['charges_1'].fillna(252,inplace=True)
```

```
#For charges_2 again i used the median value to fill the NaN value.
df['charges_2 (%)'].median() #median=10.0
df['charges_2 (%)'].fillna(10,inplace=True)
```

```
#For Minimum Price 
df['Minimum_price'].median()  #median=3157.0
df['Minimum_price'].fillna(3157,inplace=True)
```

```
#For Maximum Price
df['Maximum_price'].median() #median=5634
df['Maximum_price'].fillna(5634,inplace=True)
```

```
df.dropna(subset=['Selling_Price'],axis=0,inplace=True)
```

**Let's Encode the Categorical Features now!!**

We have two Categorical Variables to encode that are Loyalty Customer and Product Category

```
pip install category_encoders
```

```
import category_encoders as ce
# create object of Ordinalencoding
encoder1= ce.OrdinalEncoder(cols=['Loyalty_customer'],return_df=True,
                           mapping=[{'col':'Loyalty_customer',
'mapping':{'Yes':0,'No':1}}])
df_train_transformed = encoder1.fit_transform(df)
df=df_train_transformed

encoder2= ce.OrdinalEncoder(cols=['Product_Category'],return_df=True,
                           mapping=[{'col':'Product_Category',
'mapping':{'Home_decor':0,'Educational':1,'Fashion':2,'Organic':3,'Technology':4,'Hospitality':5,'Cosmetics':6,'Child_care':7,'Repair':8,'Pet_care':9}}])
df_train_transformed = encoder2.fit_transform(df)
df=df_train_transformed
df
```

```
df[(df['Selling_Price']<0)].count()
```

As we saw that there are 14 rows in the dataframe with negative Selling Price which doesn't make Sense so we will Drop these values

```
df=df[df['Selling_Price']>=0]
```

Also, I didn't take into account the **instock_date** column as i did not find any kind of dependence of the selling price/Maximum price/Minimum Price on the date and time to take this problem as a time series problem.If found please let me know.

```
X=df[['Stall_no','Market_Category','Loyalty_customer','Product_Category','Grade','Demand','Discount_avail','charges_1','charges_2 (%)','Minimum_price','Maximum_price']]
y=df['Selling_Price']
```

Now we will split the data into train and test and then we will use different ML regression based algorithms to solve this problem with best accuracy.

```
from sklearn.model_selection import train_test_split
```

```
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.30)
```

```
#Linear Regression
from sklearn.linear_model import LinearRegression
reg1=LinearRegression()
reg1.fit(X_train,y_train)
y_pred1=reg1.predict(X_test)
```

```
#Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor
reg2=DecisionTreeRegressor()
reg2.fit(X_train,y_train)
y_pred2=reg2.predict(X_test)
```

```
#Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor
reg3=RandomForestRegressor(n_estimators=100,max_depth=10,random_state=0)
reg3.fit(X_train,y_train)
y_pred3=reg3.predict(X_test)
```

```
#Support Vector Regressor
from sklearn.svm import SVR
svr=SVR(C=0.5,kernel='poly',epsilon=0.5)
svr.fit(X_train,y_train)
y_pred5=svr.predict(X_test)
```

```
#XGBoost Regressor
from xgboost import XGBRegressor
reg4=XGBRegressor()
reg4.fit(X_train,y_train)
y_pred4=reg4.predict(X_test)
```

```
from sklearn.metrics import mean_squared_error,r2_score

mse=mean_squared_error(y_test,y_pred1)
rmse1=(np.sqrt(mse))
r21=r2_score(y_test,y_pred1)

mse=mean_squared_error(y_test,y_pred2)
rmse2=(np.sqrt(mse))
r22=r2_score(y_test,y_pred2)

mse=mean_squared_error(y_test,y_pred3)
rmse3=(np.sqrt(mse))
r23=r2_score(y_test,y_pred3)

mse=mean_squared_error(y_test,y_pred4)
rmse4=(np.sqrt(mse))
r24=r2_score(y_test,y_pred4)

mse=mean_squared_error(y_test,y_pred5)
rmse5=(np.sqrt(mse))
r25=r2_score(y_test,y_pred5)

print('Linear Regression -> rmse:{}, r2_score:{}' .format(rmse1,r21))

print('Decision Tree     -> rmse:{}, r2_score:{}' .format(rmse2,r22))

print('Random Forest     -> rmse:{}, r2_score:{}' .format(rmse3,r23))

print('XG-Boost          -> rmse:{}, r2_score:{}' .format(rmse4,r24))

print('SVR               -> rmse:{}, r2_score:{}' .format(rmse5,r25))
```

As both **rmse_value** and **r2score** are highest for **Random Forest Regressor**, i will move forward with this Model and will try to improve it further and also check for overfitting using Cross Validation.

```
#Random Forest Regressor

#checking the accuracy for different hyperparameters using GridSearchCV 

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

estimator=RandomForestRegressor(random_state=0)

param_grid = { 
            "n_estimators"      : [100,200,250],
            "max_features"      : ["auto", "sqrt", "log2"],
            "max_depth" :[5,10,15],
            "bootstrap": [True, False],
            }

grid1=GridSearchCV(estimator, param_grid, n_jobs=-1, cv=5)

grid1.fit(X_train,y_train)
grid1.best_params_
```

```
reg=RandomForestRegressor(n_estimators=100,max_depth=15,max_features='auto',bootstrap=True,random_state=0)
```

```
reg.fit(X_train,y_train)
y_pred=reg.predict(X_test)
```
